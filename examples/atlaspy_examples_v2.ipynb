# Example ATLASpy use cases

```python
import atlaspy.core as apy
```

## Plotting Default Brain Atlases

ATLASpy comes with 6 default brain atlases:

- `ho`: HarvardOxford
- `aal`: AAL116
- `bna`: Brainetomme
- `sch400`: Schaefer 400
- `dkt`: Desikan-Killani-Toureville
- `vep`: Virtual Epilepsy Project sub-parcellation of DKT

There is additionally the option of adding more atlases. See below,

### Example 1

```python
# Example 1: Plotting a raw brain atlas with default settings
atlas_name = 'ho'
apy.plot_raw_atlas(atlas_name)
```

This example demonstrates how to plot a raw brain atlas using the default settings. The atlas_name variable specifies the name of the atlas to be plotted. 

- `ho`: HarvardOxford
- `aal`: AAL116
- `bna`: Brainetomme
- `sch400`: Schaefer 400
- `dkt`: Desikan-Killani-Toureville
- `vep`: Virtual Epilepsy Project sub-parcellation of DKT

In this case, the 'ho' atlas is used. The function plot_raw_atlas() automatically assigns a random color to each label in the atlas and displays the plot.

### Example 2

```python

# Example 2: Plotting a raw brain atlas with customized colormap and camera position
atlas_name = 'ho'
cmap = 'jet'
camera_position = 'right'
apy.plot_raw_atlas(atlas_name, cmap=cmap, camera_position=camera_position, interactive=True)
```

In this example, a raw brain atlas is plotted with customized settings. The cmap parameter specifies the colormap to be used for coloring the labels. Here, the `jet` colormap is chosen. The camera_position parameter determines the viewpoint for the plot. The `right` position is selected in this case. The interactive parameter is set to `True`, enabling an interactive plot that allows for navigation and exploration of the atlas.

### Example 3

```python

# Example 3: Plotting a raw brain atlas in left-right-top views
apy.plot_raw_atlas_lrt(atlas_name, cmap='Set3')

```

This example showcases how to plot a raw brain atlas in left-right-top views. The function `plot_raw_atlas_lrt()` plots the atlas from three different perspectives: left, right, and top. The `cmap` parameter specifies the colormap used to color the labels. Here, the `Set3` colormap is used to provide distinct colors for the labels.

### Example 4

```python
# Example 4: Plotting a raw brain atlas in left-right-medial views
apy.plot_raw_atlas_lrm(atlas_name, cmap='Set3')
```

In this example, a raw brain atlas is plotted in left-right-medial views. The function `plot_raw_atlas_lrm()` displays the atlas from three different perspectives: left, right, and medial (middle). Similar to the previous example, the `cmap` parameter sets the colormap for coloring the labels, and the `Set3` colormap is chosen here.

### Example 5: Plot the subcortical regions of a given brain atlas

Plot Harvard-Oxford atlas subcortical structures

```python
apy.plot_raw_atlas_subcortical_lrt('ho', cmap='Set3')
```

Plot AAL atlas subcortical structures

```python
apy.plot_raw_atlas_subcortical_lrt('aal', cmap='Set3')
```

## Plotting scalar values for each atlas' brain region

The primary use of ATLASpy is for researchers to be able to plot scalar values obtained for different brain regions on a 3D atlas plot. These examples show how to achieve that.

### Example 6: Plotting values on atlas directly from a NIFTI file

Say we have a NIFTI file. We can use ATLASpy to directly parcellate the image into an atlas and plot it. For this function to work, the nifti has to be in MNI space, specifically the tpl-MNI152NLin2009cAsym_res-01_desc-brain_T1w space (this template can be found here: https://github.com/allucas/atlaspy/tree/main/atlaspy/source_data/atlases/niftis). If the data is already in MNI space and it only needs to be resliced to this template, the option `reslice=True` can be used

```python
nifti_file = '/Users/allucas/Documents/research/CNT_staging/P50_atlaspy/code/atlaspy/python_package/test/atlaspy/atlaspy/source_data/atlases/niftis/sch400.nii.gz'
apy.plot_atlas_from_nifti_lrt('ho', nifti_file, cmap='RdYlBu', reslice=True)
```

To extract the data that is being plotted, we can use `apy.get_dataframe_from_nifti()`, the output will contain columns: `atlas_index`, which represents the value of the index in the atlas; `roi_value`, which represents the scalar that is assigned to that atlas index. Correspondance between the atlas index and the brain regions is provided by our lookup tables (https://github.com/allucas/atlaspy/tree/main/atlaspy/source_data/atlases/luts)

```python
apy.get_mean_dataframe_from_nifti('ho',nifti_file, reslice=True)
```

The same NIFTI, but visualized in the Schaefer 400 atlas

```python
apy.plot_atlas_from_nifti_lrt('sch400', nifti_file, reslice=True)
```

### Example 7: Plotting values from a CSV file or list of values

Sometimes, data is available in the form of a list of values for each brain region. This is also easily plotted in atlaspy using the `plot_rois_atlas()` and the `_lrt()` and `_lrm()` versions of the function. 

The input of these functions is a Pandas dataframe with mandatory columns `atlas_index`, which represents the index of the brain region in the atlas; and `roi_value`, which represents the scalar that we wish to visualize. We provide all of our atlases here https://github.com/allucas/atlaspy/tree/main/atlaspy/source_data/atlases/niftis, and all of the lookup tables here: https://github.com/allucas/atlaspy/tree/main/atlaspy/source_data/atlases/luts. We include an example CSV with cortical thickness values for plotting in the DKT atlas here: https://github.com/allucas/atlaspy/blob/main/examples/cortical_thickness_dkt.csv

```python
import pandas as pd
df_values = pd.read_csv('/Users/allucas/Documents/research/CNT_staging/P50_atlaspy/code/atlaspy/python_package/atlaspy/examples/cortical_thickness_dkt.csv')

# plot an interactive single view
apy.plot_rois_atlas('dkt', df_values, interactive=True)

# plot the 4 views
apy.plot_rois_atlas_lrm('dkt', df_values)

# plot the subcortical structures
apy.plot_subcortical_brain_regions_lrt('dkt', df_values)
```

## Adding a custom atlas

ATLASpy is not restricted to the default atlases included, but it can use any atlas as long as STL files are available for it. See here for instructions on how to generate these STL files: https://github.com/allucas/atlaspy/tree/main/examples/custom_atlas_readme.md

Once STL files are available, it is straightforward to add them as an atlas in ATLASpy


```python
apy.add_custom_atlas('ashsthomas',
                     '/Users/allucas/Documents/research/CNT_staging/P50_atlaspy/code/atlaspy/source_data/atlases/stls/ashsthomas_stls')
```

Plotting the newly added custom atlas

```python
atlas_name = 'ashsthomas'
cmap = 'jet'
camera_position = 'right'
apy.plot_raw_atlas(atlas_name, cmap=cmap, camera_position=camera_position, interactive=True)
```

Plotting custom values for the newly added custom atlas

```python
import pandas as pd
atlas_indices = [1,2,3,4,5,101,102,103,104,105]
values = [100,134,123,123,123,49,43,23,49,19]
df_values = pd.DataFrame()
df_values['atlas_index'] = atlas_indices
df_values['roi_value'] = values

apy.plot_rois_atlas('ashsthomas',df_values, cmap='RdYlBu' ,interactive=True)
```
